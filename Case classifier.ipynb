```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
RS=42
```

```python
#create array to append to existing dataframe to act as label
outcomes=[]
ids=[]
import random
for i in range(910):
    outcomes.append(random.randint(0,1))
    ids.append(i+1)
outcomes=np.array([ids,outcomes]).transpose()
```

```python
outcomes=pd.DataFrame(outcomes)
outcomes=outcomes.rename(columns={1:'outcome',0:'ID'})
lawyers=lawyers.merge(outcomes,how='right',on='ID')#add label data
```

```python
#suppose there is no filtering
rf=RandomForestClassifier(random_state=RS)
atts=["ID","outcome"]#attributes to remove during training
X=lawyers.drop('ID',axis=1)
X=X.drop('outcome',axis=1)
y=lawyers['outcome']
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=RS,test_size=0.5)
```
